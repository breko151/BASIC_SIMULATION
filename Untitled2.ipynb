N = 15

```python
umbral = 0.95
n = 1000
sumas = list()
for i in range(n):
    random_nums = np.random.uniform(0, 1, 15)
    calidad = [1 if i > umbral else 0 for i in random_nums]
    suma_calidad = sum(calidad)
    sumas.append(suma_calidad)
```

```python
plt.hist(sumas)
plt.show()
```

Ejercicio de Transformada inversa

```python
numeros = [0.747, 0.687, 0.303, 0.609, 0.587]
def transformada_inversa(num):
    list_values = list()
    for n in num:
        if n >= 0 and n <= 1:
            x = (4 * n - 1) ** (1 / 3)
            list_values.append(x)
    return list_values
```

```python
results = transformada_inversa(numeros)
print(results)
```

```python
plt.hist(results)
plt.show()
```

```python
numeros = [0.747, 0.687, 0.303, 0.609, 0.587]
def transformada_inversa(num):
    list_values = list()
    for n in num:
        if n >= 1/4 and n <= 1:
            x = (2/3) + (4/3) * n
        else:
            x = 4 * n
        list_values.append(x)
    return list_values
```

```python
results_numeros =  transformada_inversa(numeros)
print(results_numeros)
```

```python
plt.hist(results_numeros)
plt.show()
```
