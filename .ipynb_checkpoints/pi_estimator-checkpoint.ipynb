1. Draw a square: Draw a square and a circle inside it. The length of the side of the square must be equal to the diameter of the circle. This ensures that the circle is completely contained within the square.  

2. Generate random points: Generate a large number of random points within the square. The x coordinate and y coordinate of these points must be between 0 and the side of the square.  

3. Determine if the points are inside the circle: For each generated point, calculate the distance from the origin (0,0) to the point. If this distance is less than or equal to the radius of the circle (half the side of the square), then the point is inside the circle.  

4. Calculate the ratio: The ratio of points that fall inside the circle to the total points generated is approximately equal to the ratio of the area of the circle to the area of the square. Since the area of the square is equal to the side of the square squared, and the area of the circle is equal to Pi times the radius squared, you can use this relationship to estimate Pi:  
$$\frac{Points-circle}{Total-points} \approx \frac{\pi * (ratio)^2}{side^2}$$
$$\pi\approx4*(\frac{Points-circle}{Total-points})$$

```python
# Import libraries needed.
from simulation.generators import congruence_method
import matplotlib.pyplot as plt
import matplotlib.patches as patches
```

```python
# Initial Parameters.
N = 10000000
    # Generator 1.
seed_1 = 123456789
m_1 = 2 ** 31 - 1
a_1 = 48271
c_1 = 0
params_1 = [seed_1, a_1, c_1, m_1]
    # Generator 2.
seed_2 = 987654321
m_2 = 2 ** 32
a_2 = 1664525
c_2 = 1013904223 
params_2 = [seed_2, a_2, c_2, m_2]
```

```python
def pi_estimator(params_1, params_2, n):
    """
        With this function, we can get an estimator of pi value.
        Arguments:
            params_1: a list of values.
            params_2: a list of values.
            n: a integer value.
        Returns:
            zip(xs, ys): a zip of different list.
            zip(xs_1, ys_1): a zip of different list.
            estimator: a float value.
    """
    # The counts of point in the circle. 
    points_in_circle = 0
    # Generation of points.
    xs = congruence_method(*params_1, n)
    ys = congruence_method(*params_2, n)
    # Initialization of values.
    xs_1 = list() 
    ys_1 = list()
    # Get distance with center. 
    for x, y in zip(xs, ys):
        distance_to_origin = x ** 2 + y ** 2
        # Validation with the circle.
        if distance_to_origin <= 1:
            points_in_circle += 1
            # Points in the circle.
            xs_1.append(x)
            ys_1.append(y)
    # Estimation.
    estimator = 4 * (points_in_circle / n)
    # Return
    return [zip(xs, ys), zip(xs_1, ys_1), estimator]
```

```python
coordenates, coordenates_circle, estimator = pi_estimator(params_1, params_2, N)
```

```python
print(f'The estimation of Pi is {estimator}')
```

```python
# Coordenates
x_1, y_1 = zip(*coordenates)
# Coordenates in circle
x_2, y_2 = zip(*coordenates_circle)
```

```python
# Create fig and ax.
fig, ax = plt.subplots()

# Draw circle. 
circle = plt.Circle((0, 0), 1, color='b', fill=False)
ax.add_artist(circle)

# Draw square.
square = patches.Rectangle((-1, -1), 2, 2, linewidth=1, edgecolor='r', facecolor='none')
ax.add_patch(square)

# All points.
lenght_points_1 = len(x_1)
ax.scatter(x_1[:int(lenght_points_1 / 10000)], y_1[:int(lenght_points_1 / 10000)], color='green', label="Out Circle")

# Points in circle.
lenght_points_2 = len(x_2)
ax.scatter(x_2[:int(lenght_points_2 / 10000)], y_2[:int(lenght_points_2 / 10000)], color='blue', label="In circle")

# ax limits.
ax.set_xlim(0, 1.5)
ax.set_ylim(0, 1.5)
ax.set_aspect('equal', 'box')

# Pritn plt.
plt.legend()
plt.show()
```
