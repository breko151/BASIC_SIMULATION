```python
import numpy as np
import matplotlib.pyplot as plt

# Definir la función de distribución acumulativa (CDF)
def cdf(x):
    return 1 - np.exp(-x)  # CDF de una distribución exponencial con parámetro lambda = 1

# Función de transformada inversa para generar variables aleatorias
def inverse_transform_sampling(cdf, n_samples=1):
    u = np.random.uniform(0, 1, n_samples)  # Generar variables aleatorias uniformes
    samples = cdf_inv(u)  # Aplicar la función inversa de la CDF a las variables uniformes
    return samples

# Función inversa de la CDF
def cdf_inv(u):
    return -np.log(1 - u)  # Función inversa de la CDF de una distribución exponencial

# Número de muestras a generar
n_samples = 1000

# Generar muestras utilizando transformada inversa
samples = inverse_transform_sampling(cdf, n_samples)

# Visualizar el histograma de las muestras generadas
plt.hist(samples, bins=30, density=True, alpha=0.6, color='g', label='Histograma de muestras')
plt.xlabel('Valor de la variable aleatoria')
plt.ylabel('Densidad de probabilidad')
plt.title('Histograma de muestras generadas usando transformada inversa')
plt.legend()
plt.show()
```
