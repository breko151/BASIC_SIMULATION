```
import numpy as np
```

```
def generar_exponencial(lambds, n, umbral=0.7):
    """
    Genera n variables aleatorias exponenciales con parámetro lambda (lambd).
    """
    # Genera n variables aleatorias uniformes en el intervalo [0, 1]
    u = np.random.rand(n)
    variable_list = list()
    for u_i in u:
        if u_i <= umbral:
            variable_list.append((-1 / lambd[0]) * np.log(1 - u_i))
        else:
            variable_list.append((-1 / lambd[1]) * np.log(1 - u_i))
    
    return variable_list

# Parámetro lambda para la distribución exponencial
lambd = [1, 3]

# Número de variables aleatorias exponenciales a generar
n = 1000

# Genera las variables aleatorias exponenciales
variables_exponenciales = generar_exponencial(lambd, n)

# Imprime las primeras 10 variables aleatorias exponenciales generadas
print("Variables Aleatorias Exponenciales Generadas:")
print(variables_exponenciales[:10])
```
